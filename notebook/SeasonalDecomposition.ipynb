```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import statsmodels.api as sm

import seaborn as sns
#plt.rcParams['figure.dpi'] = 100
```

```python
historical = pd.read_csv('../data/historical.csv', parse_dates=['DATE'])
historical = historical.set_index('DATE').dropna()
historical.head()
```

```python
historical.plot()
```

```python
seasonal_decompose?
```

```python
decomposition = seasonal_decompose(historical,
                                   freq=12, 
                                   model='additive', 
                                  )
decomposition.plot()
```

```python
(decomposition.observed - decomposition.seasonal).plot()
```

```python
print(decomposition.observed.shape, decomposition.seasonal.shape, decomposition.trend.shape, decomposition.resid.shape)
```

```python
(decomposition.observed - decomposition.seasonal - decomposition.trend - decomposition.resid).dropna()
```
